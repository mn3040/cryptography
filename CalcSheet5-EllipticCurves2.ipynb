```python
# We take an integer n= 87053 = 263*331 and an elliptic curve E = EllipticCurve(IntegerModRing(n),[1,1]) and a point P = E([0,1,1]), we know is on the curve
# Here we take the curve in Z/nZ
n=87053
E = EllipticCurve(IntegerModRing(n),[1,1]);
P = E([0,1,1])
print(E, ",  P=",P)
```

```python
# We take B= e^( 1/sqrt(2) * Sqrt( log(n) log(log(n)) ) ) ~= 9, then M=LCM[1,2,3,4,5,6,7,8,9] = 223092870 = 2*3*5*7*11*13*17*19*23 = p1*p2*...*p9
# and if this B doesn't work make it bigger
# We will calculate M*P but in stages (in practice you want to do binary expansion and fast exponentiation but here the numbers are small so we do directly)
# [M]P = [p9][p8][p7][p6][p5][p4][p3][p2][p1] P
# Since the numbers are small (this won't be true for true factorization) We will take M= 50! instead, so we can easily write it in a loop here
Q = P
for i in range(1,100):
    try:
         Q = i*Q
         print(Q)
    except ZeroDivisionError as err:
         print("Failed when trying to compute the " + str(i) + "th multiple.")
         errStr = "Error: {0}".format(err)
         print( ' '.join(errStr.split(' ')[:7]) )
         break
```

```python
def f(k):
    Q=P
    for i in range(1,k):
        try:
             Q = i*Q
             print(factorial(i),"*P = ", Q)
        except ZeroDivisionError as err:
             print("Failed when trying to compute the " + str(i) + "th multiple, when getting (", i,"!) * P.")
             errStr = "Error: {0}".format(err)
             print( ' '.join(errStr.split(' ')[:7]) )
             break
        except AttributeError as err:
            print("other error: KeyError")
        except AttributeError as err:
            print("other error: AttributeError")
            
            
```

```python
f(14)
```
